```python
def numeric_columns(df):
    return df.select_dtypes(include='number').describe()
numeric_columns(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
sns.boxplot(y='test preparation course',x='math score',data=df,hue='gender',
           showmeans=True,palette='Set2')
plt.title('Hassan ')
plt.show()
```

```python
df['race/ethnicity'].value_counts()
```

```python
df['gender'].value_counts()
```

```python
sns.countplot(data=df, x='race/ethnicity',hue='lunch')
plt.show()
```

```python
sns.barplot(
    data=df,
    x='test preparation course',
    y='math score'
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    data=df,
    x='parental level of education',
    y='math score'
)

plt.xticks(rotation=30)
plt.title('Math Score by Parental Level of Education')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x='parental level of education',
    hue='gender'
)

plt.xticks(rotation=30)
plt.title('Distribution of Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Count')

plt.show()
```
